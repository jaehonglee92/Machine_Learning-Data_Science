### dictionary
* 키와 값을 갖는 데이터 구조
* 키는 내부적으로 hash값으로 저장
* 순서를 따지지 않음. 즉, 인덱스가 없음

```python
a = {'Korea': 'Seoul',
     'Canada': 'Ottawa',
     'USA': 'Washington D.C'}

print(a)

b = {}

type(a)
```

```python
print(a['Korea'])
```

```python
b = {0:1, 1:6, 7:9, 8:10}
print(b[0])
```

* 기존에 키가 존재하면, 새로운 값으로 업데이트
* 존재하지 않으면, 새로운 키, 값 생성

```python
a = {'Korea': 'Seoul',
     'Canada': 'Ottawa',
     'USA': 'Washington D.C'}

a['Japan'] = 'Tokyo'
a['Japan'] = 'Kyoto'
a['China'] = 'Beijing'

print(a)
```

* update()
    + 두 딕셔너리를 병합함
    + 겹치는 키가 있다면 parameter로 전달되는 키 값이 overwrite된다.

```python
a = {'a':1, 'b':2, 'c':3}
b = {'a':2, 'd':4, 'e':5}

a.update(b)

print(a)
```

* key 삭제
    + del 키워드 사용
    + pop 함수 이용

```python
a = {'a':1, 'b':2, 'c':3}
print(a)

a.pop('b')
print(a)

del a['c']
print(a)
```

* clear()
    + 딕셔너리의 모든 값을 초기화

```python
print(a)
a.clear()
print(a)
```

* in
    + key값 존재 확인
    + O(1) 연산 - 딕셔너리의 크기와 관계없이 항상 연산의 속도가 일정하다는 의미

```python
a = {'a':1, 'b':2, 'c':3}
print(a)

'b' in a
```

### 모든 keys, values 접근
* keys() - 키만 반환
* values() - 값만 반환
* items() - 키, 값의 튜플을 반환

```python
print(a)
print(list(a.keys()))
print(list(a.values()))

list(a.items())
```
