### 학습목표
#### 1. pandas Series 이해하기

```python
import numpy as np
import pandas as pd
```

#### Seires 값 변경

+ 추가 및 업데이트: 인덱스를 이용
+ 삭제: drop 함수 이용

```python
s = pd.Series(np.arange(100, 105), ['a', 'b', 'c', 'd', 'e'])
s
```

```python
s['a'] = 200
s
```

```python
s['k'] = 300
s
```

```python
s.drop('k', inplace=True)
```

```python
s
```

```python
s[['a', 'b']] = [300, 900]
s
```

#### slicing
+ 리스트, ndarray와 동일하게 적용

```python
s1 = pd.Series(np.arange(100, 105))
s1
```

```python
s1[1:3]
```

```python
s2 = pd.Series(np.arange(100, 105), ['a', 'b', 'c', 'd', 'e'])
s2
```

```python
s2[1:3]
```

```python
s2['c':'d']
```
